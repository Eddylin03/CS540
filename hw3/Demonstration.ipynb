```python
from scipy.linalg import eigh
import numpy as np
import matplotlib.pyplot as plt
from hw3 import load_and_center_dataset, get_covariance, get_eig, project_image, display_image, perturb_image, combine_image  # Import functions from hw3.py
```

# Load and center the dataset

```python
x = load_and_center_dataset('face_dataset.npy')
```



# Calculate the covariance matrix


```python
S = get_covariance(x)
```



# Get the top 100 eigenvalues and eigenvectors


```python
Lambda, U = get_eig(S, 100)
```


# Project an image onto the top 100 eigenvectors


```python
projected_image = project_image(x[50], U)
```



# Display the original and projected images


```python
fig, ax1, ax2 = display_image(x[50], projected_image)
plt.show()
```

# Perturb the image with sigma = 1000


```python
perturbed_image = perturb_image(x[50], U, 1000)
```


# Display the original and perturbed images


```python
fig, ax1, ax2 = display_image(x[50], perturbed_image)
plt.show()
```


# Combine two images with lambda = 0.5


```python
combined_image = combine_image(x[50], x[80], U, 0.5)

```

# Display the original and combined images

```python
fig, ax1, ax2 = display_image(x[50], combined_image)
plt.show()
```
